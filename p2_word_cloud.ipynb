```python
import matplotlib.pyplot as plt
import random


def preprocess_text(text):
    # Convert to lowercase and remove punctuation
    text = text.lower()
    cleaned_text = ''
    for char in text:
        if char.isalnum() or char.isspace():
            cleaned_text += char
    return cleaned_text


def generate_word_cloud(text):
    words = text.split()
    word_freq = {}

    for word in words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

    return word_freq


def plot_word_cloud(word_freq):
    plt.figure()

    # Sort words by frequency in descending order
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

    # Position words without overlap using a grid-based system
    grid_size = 10
    grid = []
    for _ in range(grid_size):
        row = [False] * grid_size
        grid.append(row)

    for word, freq in sorted_words:
        x, y = random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)

        # Find an empty grid position
        while grid[x][y]:
            x, y = random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)

        grid[x][y] = True

        # Plot the word
        color = (random.random(), random.random(), random.random())  # Random RGB color
        plt.text(x / grid_size, y / grid_size, word, fontsize=freq * 10,color=color)

    # plt.axis('off')
    plt.title('Word Cloud')
    plt.show()


text = "Artifical Intelleigence is interesting topic. it can do wonders. Artifical Artifical"
preprocessed_text = preprocess_text(text)
word_freq = generate_word_cloud(preprocessed_text)
plot_word_cloud(word_freq)
```
